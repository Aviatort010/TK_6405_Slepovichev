#### 3.1 Написать функцию формирования порождающей и проверочной матриц кода Хэмминга
(2^r^ - 1,2^r^ − r − 1,3) на основе параметра r,
а также таблицы синдромов для всех однократных ошибок.

```python
import numpy as np

# Так же, как и в первой ЛР, используется для "фокусов"
import itertools


def create_h_matrix(r_param):
    '''
    Функция, возвращающая **проверочную** матрицу Хемминга на основе параметра r
    :param r_param: параметр для создания матрицы
    :return: проверочная матрица Хемминга (h_matrix)
    '''
    i = 0
    h_matrix_local = np.zeros((2 ** r_param - 1, r_param), dtype=int)
    h_matrix_local[2 ** r_param - 1 - r_param: (2 ** r_param - 1)] = np.eye(r_param, dtype=int)
    all_r_words = [*itertools.product(range(2), repeat=r_param)]

    for element in all_r_words:
        if sum(element) > 1:
            h_matrix_local[i, :] = element
            i += 1

    return h_matrix_local


def create_g_matrix(r_param):
    '''
    Функция, возвращающая порождающую матрицу Хемминга на основе параметра r
    :param r_param: параметр для создания матрицы
    :return: порождающая матрица Хемминга
    '''

    n = 2 ** r_param - 1
    g_matrix_local = np.zeros((n - r_param, n), dtype=int)
    g_matrix_local[:, 0: (n - r_param)] = np.eye(n - r_param, dtype=int)
    g_matrix_local[:, n - r_param:] = h_matrix[0: (2 ** r_param - 1 - r_param), :].copy()
    return g_matrix_local


def create_syndrome_ones_table(h_matrix_local):
    '''
    Функция, возвращающая таблицу синдромов для **однократных** ошибок
    :param h_matrix_local: Проверочная матрица
    :return: таблица однократных синдромов
    '''

    syndromes_ones_local = {}
    for i in range(len(h_matrix_local)):
        syndromes_ones_local[np.array_str(h_matrix_local[i, :])] = i
    return syndromes_ones_local

# Зададим r пока здесь
r = 3

h_matrix = create_h_matrix(r)
g_matrix = create_g_matrix(r)
syndromes_ones = create_syndrome_ones_table(h_matrix)

print("h matrix = ")
print(h_matrix)
print("\n g_matrix = ")
print(g_matrix)


word = np.zeros(2** r - 1 - r, dtype=int)

key_word = (word @ g_matrix) % 2
broken_key_word = key_word.copy()
broken_key_word[0] = 1

error_index = syndromes_ones[np.array_str((broken_key_word @ h_matrix) % 2)]
repaired_word = broken_key_word.copy()
repaired_word[error_index] = (repaired_word[error_index] + 1) % 2

all(key_word == repaired_word)
```

#### 3.3 Написать функцию формирования порождающей и проверочной матриц расширенного кода Хэмминга (H*)
(2^r^ - 1,2^r^ − r − 1,3) на основе параметра r,
а также таблицы синдромов для всех однократных ошибок

```python
def create_h_matrix_extended(h_matrix_local):
    '''
    Функция, возвращающая **расширенную** проверочную матрицу Хемминга
    :param h_matrix_local: матрица Хемминга
    :return: расширенная матрица Хемминга
    '''
    h_matrix_extended_local = np.zeros((h_matrix_local.shape[0] + 1, h_matrix_local.shape[1] + 1), dtype=int)
    h_matrix_extended_local[0:h_matrix_local.shape[0], 0:h_matrix_local.shape[1]] = h_matrix_local.copy()
    h_matrix_extended_local[:, h_matrix_extended_local.shape[1] - 1] = np.ones(h_matrix_extended_local.shape[0])
    return h_matrix_extended_local


def create_g_matrix_extended(g_matrix_local):
    '''
    Функция, возвращающая **расширенную** порождающую матрицу Хемминга
    :param g_matrix_local: матрица Хемминга
    :return: расширенная матрица Хемминга
    '''
    g_matrix_extended_local = np.zeros((g_matrix_local.shape[0], g_matrix_local.shape[1] + 1), dtype=int)
    g_matrix_extended_local[:, 0:g_matrix_local.shape[1]] = g_matrix_local.copy()
    for i in range(g_matrix_extended_local.shape[0]):
        if sum(g_matrix_extended_local[i]) % 2 == 1:
            g_matrix_extended_local[i, g_matrix_extended_local.shape[1] - 1] = 1
    return g_matrix_extended_local


# Теперь по порядку
r = 2

h_matrix = create_h_matrix(r)
g_matrix = create_g_matrix(r)

h_matrix_extended = create_h_matrix_extended(h_matrix)
g_matrix_extended = create_g_matrix_extended(g_matrix)


print("h extended matrix = ")
print(h_matrix_extended)
print("\ng extended matrix = ")
print(g_matrix_extended)
```
